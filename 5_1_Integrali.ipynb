# Integrazione di una funzione



```python
# import
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np

from IPython.display import display,Math
```

```python
x = sym.symbols('x')

# a simple function
f = x

# integrazione indefinita
sym.integrate(f)
```

```python
# integrazione definita
sym.integrate(f,(x,0,1))
```

# Grafico di una funzione con sympy

```python
# plotting...

lowbnd = 2 # add later

# definiamo funzione e sua primitiva
f = x**3 / (x-lowbnd)
intf = sym.integrate(f)

# creiamo un oggetto grafico senza mostrarlo
p = sym.plotting.plot(f,show=False)
# Aggiunge al grafico precedente la curva dell'integrale definito di f(x) nell'intervallo [2.1,10] con il colore della linea rosso.
p.extend(sym.plotting.plot(intf,(x,2.1,10),show=False,line_color='r') )

p.xlim = [-10,10]
p.ylim = [-200,200]

p[0].label = '$f(x)$'
p[1].label = '$\\int f(x)dx$'
p.legend = True
p.show()

display(Math('f(x) = %s' %sym.latex(f)))
display(Math('\\int f(x) dx = %s' %sym.latex(intf)))
```

# Esercizio

```python
# Calcoliamo l'integrale indefinito della derivata della funzione originale
x = sym.symbols('x')
f = 2*x**3 + sym.sin(x)

df = sym.diff(f)

intf = sym.integrate(f)

display(Math('f(x)=%s' %sym.latex(f)))
display(Math('f\'=%s' %sym.latex(df)))
display(Math('\\int (f\')dx=%s' %sym.latex(intf)))
```

```python
# Creiamo un grafico per visualizzare le due funzioni
p = sym.plotting.plot(df,show=False)
p.extend(sym.plotting.plot(intf,show=False,line_color='r') )
p.extend(sym.plotting.plot(f,show=False,line_color='yellow') )

p.xlim = [-10,10]
p.ylim = [-200,200]

p[0].label = '$Df(x)$'
p[1].label = '$\\int f(x)dx$'
p[2].label = '$f(x)$'
p.legend = True
p.show()
```
